Se sabe que la función:

**Nota:** La derivada de dicha función es
$$
f^{\prime}(x)=\pap{5\cos(5x)-\sin(5x)}\ee^{-x}+1\;.
\tag{2}
$$

### 1.2 Graficar el método (valor 1.0)

Grafique la función $f(x)$ y los tres primeros pasos del método descrito en el punto anterior.


```python
#1.2 Gráficar el método


#Definimos la función y la primera derivada
n1,n2= 1,1.5


funp = lambda x: (1-(((n1/n2)**2)*(1-(nmp.cos(x))**2)))**(1/2)

fun1 = lambda x: ((n1*nmp.cos(x))-n2*funp(x))/((n1*nmp.cos(x))+n2*funp(x))
fun2 = lambda x: ((n2*nmp.cos(x))-n1*funp(x))/((n2*nmp.cos(x))+n1*funp(x))
fun3 = lambda x: ((2*n1*nmp.cos(x)))/((n2*nmp.cos(x))+n1*funp(x))
fun4 = lambda x: x*0-0.0045
#Definimos los puntos por graficar
x=nmp.linspace(0,nmp.pi/2,101)
ygf1=fun1(x)
ygf2=fun2(x)
ygf3=1 +fun1(x)
ygf4=fun3(x)



f= plt.figure(figsize=(20,25),dpi=100)
ax1 = f.add_subplot(321,title="",frameon=True,xlabel=r"$\dot{\theta_{i}}$ (rad)",
                    ylabel=" Coeficientes de amplitud")



ax1.plot(x,ygf1, color='blue')
ax1.plot(x,ygf2,color='red')
ax1.plot(x,ygf3,color='blue')
ax1.plot(x,ygf4,color='red')
ax1.plot(x, fun4(x), color='black', marker=',', linestyle='dashed', markersize=0.1)


ax1.legend(['$r_{\perp}$','$r_{\parallel }$','$t_{\perp}$','$t_{\parallel }$'])


ax1.set_ylim(-1,1)
ax1.set_xlim(0,nmp.pi/2)
plt.show();

```
