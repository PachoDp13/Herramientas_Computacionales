## 2. Base ortonormal en $\RR ^3$ (Valor 1.0)

Sean dos vectores $\rb _1 =\pap{1,2,1}$ y $\rb _ 2=\pap{3,4,5}$ en el espacio euclidiano $\RR ^3$. Se quiere construir una base ortonormal $\pac{\nunn _1,\nunn _2,\nunn _3}$ siguiendo el método de Grahm-Schmidt y el producto cruz de la siguiente manera:

1. El primer vector es $\nunn _1 = \frac{\rb _1}{\pav{\rb _1}}$
2. El segundo vector se define como $\nunn _2 = \frac{\rb _p}{\pav{\rb _p}}$, donde $\rb _p = \rb_2-\pap{\rb _1 \cdot \rb_2}\nunn _1$.
3. El tercer vector se define como $\nunn _3 = \nunn _1 \times \nunn _2$

Halle usando `numpy` estos tres vectores e imprima su respectivos arreglos de forma adecuada.

```python
#2. Base ortonormal en R^3
r1,r2=np.array([1,2,1]),np.array([3,4,5])
n1= r1/(np.linalg.norm(r1))
rp=r2-((sum(r1*r2))*n1)
n2= rp/(np.linalg.norm(rp))
n3= np.cross(n1,n2)
print("La base ortonormal de los vectores r1 y r2 es:")
print( "B= {",n1,",",n2,",",n3,"}")
```

## 3. Funciones de dos variables

Sean $f(x,y)=(x-3)^2+(y-3)^2-xy$, $g(x,y)=y-(x-1)^2$ dos funciones de dos variables reales. Sabemos que en el intervalo $(x,y)\in \pas{0,3}^2$ hay un único punto $(x_0,y_0)$ donde las dos funciones se anulan. Ese punto es el mínimo de la función $h(x,y)=f(x,y)^2+g(x,y)^2$.

### 3.1. Gráficas de las funciones (Valor 1.5)

En una sola figura con 4 `subplot` (arreglo de figuras $2\times 2$), grafique las cuatro funciones $f,g,h,\sqrt{h}$ de tal forma que la primera sea respectivamente una gráfica 3D de superficie (`plot_surface`), la segunda un `scatter` 3D, la tercera una densidad con `pcolormesh` y la cuarta un contorno con `contourf`. ### 3.2. Encuentre el $(x_0,y_0)$ (Valor 0.5)

Imprima adecuadamente la respuesta con un `print`.

```python
#3. Funciones de dos variables
#3.1 Gráficas de las funciones
s = np.linspace(0, 3)
x,y = np.meshgrid(s,s)
f = lambda x,y: ((x-3)**2)+((y-3)**2)-x*y
g = lambda x,y: y-((x-1)**2)
z1 = f(x,y)
z2 = g(x,y)
z3=(z1*z1)+(z2*z2)
z4=(z3)**(1/2)
fig = plt.figure(figsize=(30,20))
ax1 = fig.add_subplot(221, projection="3d",fc=(0,0,0,0))
ax2 = fig.add_subplot(222, projection="3d",fc=(0,0,0,0))
ax3 = fig.add_subplot(223, fc=(0,0,0,0),aspect='equal')
ax4 = fig.add_subplot(224, fc=(0,0,0,0),aspect='equal')

ax1.plot_surface(x, y, z1,cmap="inferno",alpha=0.9,edgecolor="none")
ax2.scatter(x, y, z2,c=z2.flatten(),cmap="viridis",alpha=0.7)
im3=ax3.pcolormesh(x,y,z3,cmap="jet")
im4=ax4.contourf(x,y,z4,levels=45,cmap="rainbow")
plt.colorbar(im3,ax=ax3)
plt.colorbar(im4,ax=ax4)

#3.2 Encuentre el (x0,y0)
i,j=np.unravel_index(np.argmin(z3),np.shape(z3))
xmin,ymin,zmin=x[i,j],y[i,j],z3[i,j]
print("El mínimo se encuentra en el punto (x,y)=({0:.2f},{1:.2f}) y vale f(x,y)={2:0.2f}".format(xmin,ymin,zmin))



```
